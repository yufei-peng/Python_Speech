# text tags

```python
f = open('autobiography_text.txt','r')
text = f.read().decode("utf-8").encode("utf-8")
```

```python
print '*'*10 + 'import jieba, 兩種段詞模式' + '*'*10, '\n'

# https://github.com/ywchiu/pytextmining/blob/master/20151013Tutorial.ipynb
# pip install jieba
import jieba

seg_list = jieba.cut(text, cut_all=True)
print "Full Mode:", "/ ".join(seg_list) 

seg_list = jieba.cut(text, cut_all=False)
print "Default Mode:", "/ ".join(seg_list) 
```

```python
print '*'*10 + '加入字定義的詞' + '*'*10, '\n'

jieba.add_word('明尼蘇達州立大學')
jieba.add_word('生物晶片')
jieba.add_word('MongoDB')
jieba.add_word('BigData')
jieba.add_word('資料庫')
jieba.add_word('鍾景光教授')
words = jieba.cut(text)
print "/ ".join(words)


print '*'*10 +'計算詞出現的的次數' + '*'*10, '\n'

worddic = {}
for wd in jieba.cut(text):
    if wd not in worddic:
        worddic[wd] = 1
    else:
        worddic[wd] = worddic[wd] + 1

print worddic

        
print '*'*10 +'文字雲就出來了' + '*'*10, '\n'
```
