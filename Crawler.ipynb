```python
import requests as r
from bs4 import BeautifulSoup
```

```python
# 爬取Dcard，使用網址列上的網址，做爬取，爬取最新文章的部分
# res = r.get('https://www.dcard.tw/f/sex?latest=true')
res = r.get('https://www.dcard.tw/f/photography?latest=true')
```

```python
# 將語系設定為utf-8
res.encoding='utf-8'
```

```python
# 觀察response
res
```

```python
# 觀察爬回來的是什麼
res.text
```

```python
# 用beautifulsoup做剖析
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# 觀察剖析後的檔案
soup
```

```python
# Dcard都是把影像放在 class=PostEntry_image_1E6Mi 裡面，擷取這個class
# 拿到這個class後，要把style裡的網址拿出來，取第21-55字元，為完整網址，將網址用images這個list存起來
imageblocks = soup.select('.PostEntry_image_1E6Mi')
images = []
for imageblock in imageblocks:
    imageurl = imageblock['style'][21:56]
#     print(imageurl)
    images.append(imageurl)
    images
```

```python
print(images)
print(len(images))
```

```python
from IPython.display import Image, display
from IPython.core.display import HTML
```

```python
# 用Image和display這兩個套件將圖片顯示在畫面上
pics = [display(Image(url = image, width=200)) for image in images]
```

```python
# 透過 network，觀察到Dcard傳輸後續網頁時的pattern，使用下面的網址來抓
# 此網址的回傳值response是json
res2 = r.get('https://www.dcard.tw/_api/forums/photography/posts?popular=false&before=227501912')
```

```python
# 所以不用beautifulsoup來做剖析，將檔案視為json來讀取
allData = res2.json()
len(allData)
```

```python
# 觀察一個json，可以看到圖片的url都是放在media這個key裡面，所以把這個key拿出來，就可以拿到該篇文章的所有圖片
allData[0]
```

```python
# 將每一篇文章內的圖片顯示在畫面上
for data in allData:
    print(data['id'])
    media = data['media']
    for urlJson in media:
        display(Image(url = urlJson['url'], width=200))
    print("=" * 50)
```
